```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'m')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.plot(y,x, 'g*-')
```

```python
fig = plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y, 'm')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.2, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])


axes1.plot(x,y, 'b')
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Axes 1 title')

axes2.plot(y,x, 'r')
axes2.set_xlabel('X axis2')
axes2.set_ylabel('Y axis2')
axes2.set_title('Axes 2 Title')
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title');
```

```python
fig.savefig("filename.png")
```

```python
fig.savefig("filename.png", dpi=150)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```
